```python
isConnected = False

''' 01: read in mongoDB '''
try: 
    conn = pymongo.MongoClient()
    db = conn.hackprinceton
    print "Connected to the hackprinceton server"
    isConnected = True
except:
    print "Connection to mongo server failed."
    
if isConnected:
    # connect to the main collection of the hackprinceton db
    coll = pymongo.collection.Collection(db, 'hackprinceton')
    
    # loop through each user
    userList = coll.find().distinct("userID")
    
    for user in userList:
#         properties = coll.find({"userID":user}).sort('features.properties.time')#.distinct("features.properties")
        properties = coll.aggregate([
                {"$match":
                    {"userID":user}
                },
                {"$group":
                    {"_id":    "$features.properties.time",
                     "playerid":  {"$addToSet": "$userID"},
                     "heartrate" :    {"$push": "$features.properties.heartrate"},
                     "speed":  {"$push": "$features.properties.speed"},
                     "latitude":    {"$push": "$features.properties.latitude"},
                     "longitude":    {"$push": "$features.properties.longitude"}
                    }   
                },
                {"$sort": {"_id":1}}
            ])

        # convert to list
        properties = list(properties)
        
        # get the first/second half
        firsthalf = properties[0]
        secondhalf = properties[1]
        
        # first half info
        firsttime = firsthalf["_id"]
        firstheartrate = firsthalf["heartrate"][0]
        firstspeed = firsthalf["speed"][0]
        firstlatitude = firsthalf["latitude"][0]
        firstlongitude = firsthalf["longitude"][0]

        # second half info
        secondtime = secondhalf["_id"]
        secondheartrate = secondhalf["heartrate"][0]
        secondspeed = secondhalf["speed"][0]
        secondlatitude = secondhalf["latitude"][0]
        secondlongitude = secondhalf["longitude"][0]
        
        ''' 02: Insertion into Collection '''
        # collection name
        firstcollname = str(user) + 'firsthalf'
        secondcollname = str(user) + 'secondhalf'
 
        try:
            db.create_collection(firstcollname)
            db.create_collection(secondcollname)
            
            print "Creating collection: ", firstcollname, ' --> Ok'
            print "Creating collection: ", secondcollname, ' --> Ok'
        except:
            print "Errors in creating collection: ", firstcollname
            print "Errors in creating collection: ", secondcollname
           
        # get to the collection and insert
        try:
            firstcoll = pymongo.collection.Collection(db, secondcollname)
            secondcoll = pymongo.collection.Collection(db, firstcollname)
            
            # insert for firsthalf collection
            for i in range(0, len(firstheartrate)):
                firstdata = {"time": firsttime[i],
                             "heartrate": firstheartrate[i],
                             "speed": firstspeed[i],
                             "latitude": firstlatitude[i],
                             "longitude": firstlongitude[i]
                        }
                firstjson = json.loads(json.dumps((firstdata)))
                firstcoll.insert_one(firstjson)
                
            # insert for secondhalf collection
            for i in range(0, len(secondheartrate)):
                seconddata = {"time": secondtime[i],
                                   "heartrate": secondheartrate[i],
                                   "speed": secondspeed[i],
                                   "latitude": secondlatitude[i],
                                   "longitude": secondlongitude[i]
                        }
                secondjson = json.loads(json.dumps((seconddata)))
                secondcoll.insert_one(secondjson)
        
        except:
            traceback.print_exc()
            print "errors in insertion"

    # loop through and analyze
#     for i in range(0, len(data["features"])):
    
```
